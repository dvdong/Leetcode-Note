## [389. Find the Difference (easy)](https://leetcode-cn.com/problems/find-the-difference/)

Given two strings ***s*** and ***t*** which consist of only lowercase letters.

String ***t*** is generated by random shuffling string ***s*** and then add one more letter at a random position.

Find the letter that was added in t.

````java
// 位运算：相同数异或之后结果为0，0和任意数异或为该数本身
public char findTheDifference(String s, String t) {
    int res = 0;
    for(int i = 0; i < s.length(); i++){
        res ^= s.charAt(i) ^ t.charAt(i);
    }
    res ^= t.charAt(t.length() - 1);
    return (char) res;
}

// 哈希表统计：因为都是小写字母，所以可用数组代替哈希表
public char findTheDifference2(String s, String t) {
    int[] counter = new int[26];
    for(int i = 0; i < s.length(); i++){
        char c = s.charAt(i);
        counter[c - 'a']++;
    }
    char res = ' ';
    for(int i = 0; i < t.length(); i++){
        char c = t.charAt(i);
        if(--counter[c - 'a'] < 0){
            res = c;
            break;
        }
    }
    return res;
}

// 将t的所有字符加起来减去s的所有字符加起来，剩下的就是增加的字符
public char findTheDifference(String s, String t) {
    int sum = 0;
    for(int i = 0; i < s.length(); i++){
        sum += t.charAt(i);
        sum -= s.charAt(i);
    }
    sum += t.charAt(t.length() - 1);
    return (char) sum;
}
````

## [136. Single Number (easy)](https://leetcode-cn.com/problems/single-number/)

Given a **non-empty** array of integers, every element appears *twice* except for one. Find that single one.

**Note:**

Your algorithm should have a linear runtime complexity. Could you implement it without using extra memory?

<font color='dd0000'>只有一个元素是出现一次，其他元素都出现两次，所以将数组中所有元素做异或，最后的结果就是只出现一次的元素</font>

````java
public int singleNumber(int[] nums) {
     int res = 0;
     for(int num : nums){
         res ^= num;
     }
     return res;
}
````

## [面试题56 - I. 数组中数字出现的次数 (medium)](https://leetcode-cn.com/problems/shu-zu-zhong-shu-zi-chu-xian-de-ci-shu-lcof/)

一个整型数组 ```nums``` 里除两个数字之外，其他数字都出现了两次。请写程序找出这两个只出现一次的数字。要求时间复杂度是O(n)，空间复杂度是O(1)。

<font color='dd0000'>
    
和上题不同处在于，只出现一次的数字有两个，所以全部异或只能得到这两个数字的异或值，而无法得到这两个值本身

所以，需要想办法将这两个数字分成不同的两组，并且这两组中都是除了它们本身其他数字都出现两次(即两组上题的情况)

通过 ```&``` 运算来判断一位数字不同即可分为两组，两个不同的数字至少也有一位不同,我们只需要找出那位不同的数字mask，即可完成分组（```& mask```）操作

而这两个不同的数字异或的结果恰好就反映了这两个数不同位的情况，所以只要找到它们异或结果中第一个为```1```的位置，就可以把这个数作为分组的 ```mask```

因为每个数都和这个```mask```做```&```运算，所以相同的数字一定会分到同一个组，而要求的这两个数字一定在不同组，再分别对两组数异或即可。

</font>

````java
public int[] singleNumbers(int[] nums) {
    int sum = 0;
    for(int num : nums){
        sum ^= num;
    }
    // 求异或结果sum中第一个为1的位
    int mask = 1;  // 最开始设为最低位，不断左移，直到遇到sum为1的位
    while((mask & sum) == 0){
        mask <<= 1;
    }
    int a = 0, b = 0; // a、b分别表示两组异或的结果，即待求的两个数
    for(int num : nums){
        if((num & mask) == 1){
            a ^= num;
        } else {
            b ^= num;
        }
    }
    return new int[]{a, b};
}
````
